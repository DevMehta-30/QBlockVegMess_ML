```python
from transformers import MBartForConditionalGeneration , MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
```

```python
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt",src_large="en_XX")
```

```python
input_text = ['The rain condition for this crop ideally should be 12 inches.']
```

```python
model_input = tokenizer(input_text,return_tensors='pt',padding=True,truncation=True)
```

```python
generated_tokens = model.generate(
    **model_input,
    forced_bos_token_id = tokenizer.lang_code_to_id['hi_IN']
)
```

```python
translate = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```
